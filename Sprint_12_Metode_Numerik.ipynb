### Perbaikan Data

#### Drop

```python
#Drop Duplicates
df.drop_duplicates(inplace = True)
```

```python
# Drop Features
drop_col = ['DateCrawled','DateCreated','LastSeen','NumberOfPictures','PostalCode']
df_col_drop = df.drop(columns=drop_col,axis=1)
```

```python
df_col_drop.shape
```

```python
df_col_drop.columns
```

#### missing value

```python
# VehicleType
df_col_drop['VehicleType'].fillna('other',inplace=True)
```

```python
#Power
df_col_drop.drop(df_col_drop[df_col_drop['Power'] > 500].index,inplace=True)
```

```python
#RegistrationYear
df_col_drop.drop(df_col_drop[
    (df_col_drop['RegistrationYear'] > 2022) |
    (df_col_drop['RegistrationYear'] < 1950)].index,inplace=True)
```

```python
#FuelType
df_col_drop['FuelType'].fillna('other',inplace=True)
```

```python
#Fungsi memasukan nilai yang hilang pada kolom Gearbox dengan modus terhadap kolom yang diinginkan
def fill_missing (data, column, value_column):
    grouped_values = data.groupby(column)[value_column].agg(lambda x: pd.Series.mode(x)[0]).reset_index()
    size = len(grouped_values)
    for i in range(size):
        group = grouped_values[column][i]
        value = grouped_values[value_column][i]
        data.loc[(data[column]==group) & (data[value_column].isna()),value_column] = value
    return data
```

```python
#Gearbox
#Mengisi nilai yang hilang di Gearbox dengan model dan Brand
df_col_drop = fill_missing (df_col_drop,'Model','Gearbox')
df_col_drop = fill_missing (df_col_drop,'Brand','Gearbox')
```

```python
df_col_drop['Gearbox'].value_counts(dropna=False)
```

```python
#NotRepaired
df_col_drop['NotRepaired'].fillna('unknown',inplace=True)
```

```python
df_col_drop['Model'].fillna('other',inplace=True)
```

```python
df_col_drop.isna().sum()
```

```python
#Reset index
df_col_drop.reset_index(drop=True,inplace=True)
```

#### Pemisahan data

```python
df_col_drop.shape
```

```python
# Data untuk Linear Regression, Random Forest, dan TreeClassifier
data_ohe = df_col_drop.drop(columns = 'Model',axis = 1 )
data_ohe = pd.get_dummies(data_ohe)
data_ohe.shape
```

```python
data_category = df_col_drop.copy()
data_category.shape
```

```python
category_columns = ['VehicleType','Gearbox','Model','FuelType','Brand','NotRepaired']

data_category[category_columns] = data_category[category_columns].astype('category')
```

## Pelatihan Model

### Train test split data

#### Data OHE

```python
#Data OHE
data_ohe_train,data_ohe_valid = train_test_split(data_ohe,test_size=0.3,random_state=12345)
data_ohe_valid,data_ohe_test = train_test_split(data_ohe_valid,test_size=0.5,random_state=12345)
print(data_ohe_train.shape)
print(data_ohe_valid.shape)
print(data_ohe_test.shape)
```

```python
features_train =data_ohe_train.drop(['Price'],axis=1)
target_train = data_ohe_train['Price']

features_valid = data_ohe_valid.drop(['Price'],axis=1)
target_valid = data_ohe_valid['Price']

features_test = data_ohe_test.drop(['Price'],axis=1)
target_test = data_ohe_test['Price']
```

#### Data Category

```python
data_cat_train,data_cat_valid = train_test_split(data_category,test_size=0.3,random_state=12345)
data_cat_valid,data_cat_test = train_test_split(data_cat_valid,test_size=0.5,random_state=12345)
print(data_cat_train.shape)
print(data_cat_valid.shape)
print(data_cat_test.shape)
```

```python
features_train_cat =data_cat_train.drop(['Price'],axis=1)
target_train_cat = data_cat_train['Price']

features_valid_cat = data_cat_valid.drop(['Price'],axis=1)
target_valid_cat = data_cat_valid['Price']

features_test_cat = data_cat_test.drop(['Price'],axis=1)
target_test_cat = data_cat_test['Price']
```

### Sanity Check

```python
model = LinearRegression()
```

```python
%%time

model.fit(features_train,target_train)
```

```python
%%time

pred_train = model.predict(features_train)
pred_valid = model.predict(features_valid)
pred_test = model.predict(features_test)

print('RMSE train =', mean_squared_error(target_train,pred_train)** 0.5)
print('RMSE valid =', mean_squared_error(target_valid,pred_valid)** 0.5)
print('RMSE test =', mean_squared_error(target_test,pred_test)** 0.5)
```

### Fungsi check hasil RMSE

```python
# Data OHE
def hasil_rmse(model):
    pred_train = model.predict(features_train)
    pred_valid = model.predict(features_valid)
    pred_test = model.predict(features_test)

    print('RMSE train =', mean_squared_error(target_train,pred_train)** 0.5)
    print('RMSE valid =', mean_squared_error(target_valid,pred_valid)** 0.5)
    print('RMSE test =', mean_squared_error(target_test,pred_test)** 0.5)
```

```python
# Data Category
def hasil_rmse_cat(model):
    pred_train_cat = model.predict(features_train_cat)
    pred_valid_cat = model.predict(features_valid_cat)
    pred_test_cat = model.predict(features_test_cat)

    print('RMSE train =', mean_squared_error(target_train_cat,pred_train_cat)** 0.5)
    print('RMSE valid =', mean_squared_error(target_valid_cat,pred_valid_cat)** 0.5)
    print('RMSE test =', mean_squared_error(target_test_cat,pred_test_cat)** 0.5)
```
