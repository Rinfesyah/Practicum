## Tuning Hyperparameter

### Max Depth

```python
for depth in range(1,10):
    modelRF_2 =RandomForestClassifier(random_state=12345,max_depth=depth)
    modelRF_2.fit(features_train1,target_train1)
    predict_valid6 = modelRF_2.predict(features_valid1)
    result =accuracy_score(target_valid1,predict_valid6)

    predict_test7 = modelRF_2.predict(features_test1)
    result2 = f1_score(target_test1,predict_test7)
    print('-----------------------------------------------------------')
    print (f'Hasil akurasi validation = {result*100},F1 skor = {result2}, max_depth={depth} ')
```

### n_estimators

```python
for est in range(9,20):
    modelRF_3 =RandomForestClassifier(random_state=12345,max_depth=9,n_estimators= est)
    modelRF_3.fit(features_train1,target_train1)
    predict_valid6 = modelRF_3.predict(features_valid1)
    result3 =accuracy_score(target_valid1,predict_valid6)

    predict_test8 = modelRF_3.predict(features_test1)
    result4 = f1_score(target_test1,predict_test8)
    print('-----------------------------------------------------------')
    print (f'Hasil akurasi validation = {result3*100},F1 skor = {result4} ,n_est={est} ')
```

### Class Weight

```python
class_weight = ['balanced','balanced_subsample',None]
for weight in class_weight:
    modelRF_4 =RandomForestClassifier(random_state=12345,max_depth=9,n_estimators=15,class_weight=weight)
    modelRF_4.fit(features_train1,target_train1)
    predict_valid7 = modelRF_4.predict(features_valid1)
    result5 = accuracy_score(target_valid1,predict_valid7)

    predict_test9 = modelRF_4.predict(features_test1)
    result6 = f1_score(target_test1,predict_test9)
    print('-----------------------------------------------------------')
    print (f'Hasil akurasi validation = {result5*100},F1 skor = {result6} ,class_weight={weight} ')
```

## Data Balance

```python
target_train1.value_counts()
```

```python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)

    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345
    )

    return features_upsampled, target_upsampled
```

```python
features_upsampled, target_upsampled = upsample(
    features_train1, target_train1, 3
)
```

```python
target_upsampled.value_counts()
```

```python
#fungsi untuk membuat model dari data 1
def machine_learning3(modeltype,features,target):
    #model
    model = modeltype
    model.fit(features,target)
    #predict
    predict_train = model.predict(features)
    predict_valid = model.predict(features_valid1)
    predict_test = model.predict(features_test1)
    #proba
    probabilities_valid = model.predict_proba(features_valid1)
    probabilities_one_valid = probabilities_valid[:, 1]
    #print
    print(f'akurasi train =  {accuracy_score(target,predict_train)}')
    print(f'akurasi validation =  {accuracy_score(target_valid1,predict_valid)}')
    print(f'akurasi test =  {accuracy_score(target_test1,predict_test)}')
    print(f'F1 skor =  {f1_score(target_test1,predict_test)}')
    print(f'auc_roc skor =  {roc_auc_score(target_valid1,probabilities_one_valid)}')
```

```python
machine_learning3(RandomForestClassifier(random_state=12345,max_depth=9,n_estimators=15,class_weight='balanced_subsample'),
                  features_upsampled,
                  target_upsampled)
```

### Downsampling

```python
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)]
        + [features_ones]
    )
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)]
        + [target_ones]
    )

    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345
    )

    return features_downsampled, target_downsampled
```

```python
target_train1.value_counts()
```

```python
features_downsampled, target_downsampled = downsample(
    features_train1, target_train1, 0.3
)
```

```python
target_downsampled.value_counts()
```

```python
machine_learning3(RandomForestClassifier(random_state=12345,max_depth=9,n_estimators=15,class_weight='balanced_subsample'),
                  features_downsampled,
                  target_downsampled)
```

KESIMPULAN
1. dari data balance Upsample mengalami peningkatan pada F1 skor tetapi mengalami penurunan pada akurasi validation
2. hasil f1 skor pada dataset upsample lebih baik dan melebihi target .
3. selanjut nya kita akan gunakan model yang sudah di upsample sebagai final model dengan tuning
