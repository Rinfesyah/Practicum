## Pisah data

```python
features = df.drop(columns='is_ultra',axis=1)
target = df['is_ultra']
```

```python
feature_train,feature_valid,target_train,target_valid = train_test_split(features,target,test_size=0.2,random_state=12345)
```

```python
feature_train,feature_test, target_train,target_test = train_test_split(feature_train,target_train,
                                                                        test_size=0.25,random_state=12345)
```

```python
print('Besaran data untuk Training Set =',feature_train.shape,target_train.shape)
print('Besaran data untuk Validation Set =',feature_valid.shape,target_valid.shape)
print('Besaran data untuk Test Set =',feature_test.shape,target_test.shape)
```

**KESIMPULAN**

Total data konsumen yang kita punya adalah **3214**, dikarenakan kita hanya memiliki 1 dataframe, oleh karena itu diperlukan **Split/Pembagian** data menjadi data **Training**, data **Validation** dan data **Test** dengan proporsi data sebesar **60:20:20**, 60 untuk training, 20 untuk validation dan 20 untuk test.

## Kualitas Model

### DecisionTreeClassifier

#### Tanpa Tuning Hyperparameter

```python
model_tree = DecisionTreeClassifier(random_state=12345)
model_tree.fit(feature_train,target_train)

predict_tree_valid = model_tree.predict(feature_valid)
result_tree_valid = accuracy_score(target_valid,predict_tree_valid)

predict_tree_test = model_tree.predict(feature_test)
result_tree_test = accuracy_score(target_test,predict_tree_test)

print (f'Hasil akurasi validation = {result_tree_valid*100}, Hasil akurasi test = {result_tree_test*100}')
```

Hasil dari Model DecisionTreeClassifier tanpa tuning Hyperparameter didapat model mengalami Underfitting/ penurunan hasil.

#### Tuning Hyperparameter

##### Max Depth

```python

for depth in range(1,20):
    model_tree_1 = DecisionTreeClassifier(random_state=12345,max_depth=depth)
    model_tree_1.fit(feature_train,target_train)

    predict_tree_valid_1 = model_tree_1.predict(feature_valid)
    result_tree_valid_1 = accuracy_score(target_valid,predict_tree_valid_1)

    predict_tree_test_1 = model_tree_1.predict(feature_test)
    result_tree_test_1 = accuracy_score(target_test,predict_tree_test_1)


    print('-----------------------------------------------------------')
    print (f'Hasil akurasi validation = {result_tree_valid_1*100}, Hasil akurasi test = {result_tree_test_1*100}, max_depth ={depth}')

```

setelah tuning hyperparameter max_depth dilakukan, max_depth=13 sudah mendapatkan hasil yang sama antara valid dan test, oleh karena itu kita akan mengambil best max_depth=14, karena hasil test lebih besar dari hasil valid.

##### Criterion

```python
criterion = ['gini','entropy']
for crit in criterion:
    model_tree_2 = DecisionTreeClassifier(random_state=12345,max_depth=14,criterion=crit)
    model_tree_2.fit(feature_train,target_train)

    predict_tree_valid_2 = model_tree_2.predict(feature_valid)
    result_tree_valid_2 = accuracy_score(target_valid,predict_tree_valid_2)

    predict_tree_test_2 = model_tree_2.predict(feature_test)
    result_tree_test_2 = accuracy_score(target_test,predict_tree_test_2)

    print (f'Hasil akurasi validation = {result_tree_valid_2*100}, Hasil akurasi test = {result_tree_test_2*100},criterion = {crit}')
```

Saya tidak tau kenapa Hyperparameter criterion=log_loss malah menampilkan error, sedangkan gini dan entropy tidak, jika di lihat dari hasil hyperparam criterion= gini /default lebih bagus.

##### Splitter

```python
splitter = ['best','random']
for split in splitter:
    model_tree_4 = DecisionTreeClassifier(random_state=12345,max_depth=14,criterion='gini',splitter=split)
    model_tree_4.fit(feature_train,target_train)

    predict_tree_valid_4 = model_tree_4.predict(feature_valid)
    result_tree_valid_4 = accuracy_score(target_valid,predict_tree_valid_4)

    predict_tree_test_4 = model_tree_4.predict(feature_test)
    result_tree_test_4 = accuracy_score(target_test,predict_tree_test_4)

    print (f'Hasil akurasi validation = {result_tree_valid_4*100}, Hasil akurasi test = {result_tree_test_4*100},splitter = {split}')
```

Hyperparameter splitter=best /default sudah cukup bagus menampilkan hasil akurasi test lebih tinggi dari valid

##### max_features

```python
max_features = ['auto', 'sqrt', 'log2']
for features in max_features:
    model_tree_3 = DecisionTreeClassifier(random_state=12345,max_depth=14,criterion='gini',splitter='best',max_features=features)
    model_tree_3.fit(feature_train,target_train)

    predict_tree_valid_3 = model_tree_3.predict(feature_valid)
    result_tree_valid_3 = accuracy_score(target_valid,predict_tree_valid_3)

    predict_tree_test_3 = model_tree_3.predict(feature_test)
    result_tree_test_3 = accuracy_score(target_test,predict_tree_test_3)

    print (f'Hasil akurasi validation = {result_tree_valid_3*100}, Hasil akurasi test = {result_tree_test_3*100},max_features = {features}')
```

dengan tuning hyperparameter max_feature, membuat hasil valid dan test menurun

#### Kesimpulan

model DecisionTreeClassifier dengan tuning Hyperparameter, random_state=12345, max_depth =14, criterion=gini, splitter=best.
memiliki hasil paling bagus dari yang lain nya

```python
model_tree_5 = DecisionTreeClassifier(random_state=12345,max_depth=14,criterion='gini',splitter='best')
model_tree_5.fit(feature_train,target_train)

predict_tree_valid_5 = model_tree_5.predict(feature_valid)
result_tree_valid_5 = accuracy_score(target_valid,predict_tree_valid_5)

predict_tree_test_5 = model_tree_5.predict(feature_test)
result_tree_test_5 = accuracy_score(target_test,predict_tree_test_5)

print (f'Hasil akurasi validation = {result_tree_valid_5*100}, Hasil akurasi test = {result_tree_test_5*100}')
```

### RandomForestClassifier

#### Tanpa Tuning Hyperparameter

```python
model_forest = RandomForestClassifier(random_state=12345)
model_forest.fit(feature_train,target_train)

predict_forest_valid = model_forest.predict(feature_valid)
result_forest_valid = accuracy_score(target_valid,predict_forest_valid)

predict_forest_test = model_forest.predict(feature_test)
result_forest_test = accuracy_score(target_test,predict_forest_test)

print (f'Hasil akurasi validation = {result_forest_valid*100}, Hasil akurasi test = {result_forest_test*100}')
```

dari hasil yang didapat tanpa tuning, sudah cukup bagus, model tidak underfitting maupun overfitting, tapi kita coba tuning

#### Tuning Hyperparameter

##### n_estimators

```python
n_estimators = [ 10,50,100,200,250]
for est in n_estimators:
    model_forest_1 = RandomForestClassifier(random_state=12345,n_estimators=est)
    model_forest_1.fit(feature_train,target_train)

    predict_forest_valid_1 = model_forest_1.predict(feature_valid)
    result_forest_valid_1 = accuracy_score(target_valid,predict_forest_valid_1)

    predict_forest_test_1 = model_forest_1.predict(feature_test)
    result_forest_test_1 = accuracy_score(target_test,predict_forest_test_1)

    print('-----------------------------------------------------------')
    print (f'Hasil akurasi validation = {result_forest_valid_1*100}, Hasil akurasi test = {result_forest_test_1*100} n_estimators={est} ')
```

setelah mencoba tuning n_estimator lebih dari 250, hasil test semakin menurun dan kecepatan eksekusi menjadi semakin lambat. jadi hasil terbaik tuning n_estimator adalah 200

##### criterion

```python
#var criterion sudah ada di model DecissionTreeClassifier
for crit in criterion:
    model_forest_2 = RandomForestClassifier(random_state=12345,n_estimators=200,criterion=crit)
    model_forest_2.fit(feature_train,target_train)

    predict_forest_valid_2 = model_forest_2.predict(feature_valid)
    result_forest_valid_2 = accuracy_score(target_valid,predict_forest_valid_2)

    predict_forest_test_2 = model_forest_2.predict(feature_test)
    result_forest_test_2 = accuracy_score(target_test,predict_forest_test_2)

    print('-----------------------------------------------------------')
    print (f'Hasil akurasi validation = {result_forest_valid_2*100}, Hasil akurasi test = {result_forest_test_2*100} criterion={crit} ')
```

tuning entropy membuat hasil valid dan test mengalami peningkatan

##### Max_depth

```python
for depth in range(1,6):
    model_forest_3 = RandomForestClassifier(random_state=12345,n_estimators=200,criterion='entropy',max_depth=depth)
    model_forest_3.fit(feature_train,target_train)

    predict_forest_valid_3 = model_forest_3.predict(feature_valid)
    result_forest_valid_3 = accuracy_score(target_valid,predict_forest_valid_3)

    predict_forest_test_3 = model_forest_3.predict(feature_test)
    result_forest_test_3 = accuracy_score(target_test,predict_forest_test_3)

    print('-----------------------------------------------------------')
    print (f'Hasil akurasi validation = {result_forest_valid_3*100}, Hasil akurasi test = {result_forest_test_3*100} max_depth={depth} ')
```

jika tuning max_depth pada model random forest membuat hasil test mengalami penurunan, meskipun max_depth diperbesar, jikan dipaksakan malah membuat proses kecepatan program melambat

##### max_features

```python
for features in max_features:
    model_forest_4 = RandomForestClassifier(random_state=12345,n_estimators=200,criterion='entropy',max_features=features)
    model_forest_4.fit(feature_train,target_train)

    predict_forest_valid_4 = model_forest_4.predict(feature_valid)
    result_forest_valid_4 = accuracy_score(target_valid,predict_forest_valid_4)

    predict_forest_test_4 = model_forest_4.predict(feature_test)
    result_forest_test_4 = accuracy_score(target_test,predict_forest_test_4)

    print('-----------------------------------------------------------')
    print (f'Hasil akurasi validation = {result_forest_valid_4*100}, Hasil akurasi test = {result_forest_test_4*100} max_features={features} ')
```

Tidak ada perubahan yang terjadi setelah tuning max_feature, karna tidak ada perubahan selanjutnya akan dibuat default

tuning class_weight=None lebih baik dari yang balanced dan balanced_subsample

### Logistic Regression

#### Tanpa Tuning Hyperparameter

```python
model_log = LogisticRegression(random_state=12345)
model_log.fit(feature_train,target_train)

predict_log_train = model_log.predict(feature_train)
result_log_train = accuracy_score(target_train,predict_log_train)

predict_log_valid = model_log.predict(feature_valid)
result_log_valid = accuracy_score(target_valid,predict_log_valid)

print (f'Hasil akurasi train = {result_log_train*100}, Hasil akurasi validation = {result_log_valid*100}')
```

dari hasil model tanpa tuning sudah cukup bagus diatas 75%

#### Tuning Hyperparameter

##### solver

```python
solver = ['lbfgs', 'liblinear', 'newton-cg', 'sag', 'saga']
for sol in solver:
    model_log_1 = LogisticRegression(random_state=12345,solver=sol)
    model_log_1.fit(feature_train,target_train)

    predict_log_train_1 = model_log_1.predict(feature_train)
    result_log_train_1 = accuracy_score(target_train,predict_log_train_1)

    predict_log_valid_1 = model_log_1.predict(feature_valid)
    result_log_valid_1 = accuracy_score(target_valid,predict_log_valid_1)

    print (f'Hasil akurasi train = {result_log_train_1*100}, Hasil akurasi validation = {result_log_valid_1*100}, solver ={sol}')
```

tuning solver=newton-cg dan lbfgs membuat akurasi meningkat, dan tuning sag/saga malah membuat akurasi menurun

##### penalty

```python
penalty = ['l1','l2']
for penal in penalty:
    model_log_2 = LogisticRegression(random_state=12345,solver='liblinear' ,penalty=penal)
    model_log_2.fit(feature_train,target_train)

    predict_log_train_2 = model_log_2.predict(feature_train)
    result_log_train_2 = accuracy_score(target_train,predict_log_train_2)

    predict_log_valid_2 = model_log_2.predict(feature_valid)
    result_log_valid_2 = accuracy_score(target_valid,predict_log_valid_2)

    print (f'Hasil akurasi train = {result_log_train_2*100}, Hasil akurasi validation = {result_log_valid_2*100}, penalty ={penal}')
```

dari hasil tuning penalty di dapat
solver = liblinear hanya bisa menampung penalty= l1 dan l2
solver = lbfgs dan newton-cg hanya bisa menampung penalty = l2 dan 'none'
solver = sag/saga hanya bisa menampung penalty= elasticnet

hasil solver liblinear dan penalty l1 akan menjadi pilihan untuk diteruskan tuning

##### multi_class
